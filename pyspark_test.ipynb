```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Test").getOrCreate()
df = spark.read.option("encoding", "cp949").csv("kosdaq_list.csv", header=True, inferSchema=True)

df.show()
```

```python
df2 = df.select(["종목코드","기업명","상장주식수(주)","자본금(원)","액면가(원)"])
df2 = df2.withColumnRenamed("종목코드", "code").withColumnRenamed("기업명", "name").withColumnRenamed("상장주식수(주)", "stock_count").withColumnRenamed("자본금(원)", "capital").withColumnRenamed("액면가(원)", "face_value")
df2.show()
```
